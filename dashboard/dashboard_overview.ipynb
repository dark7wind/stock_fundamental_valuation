```python
# config
pn.extension()
@pn.depends(pn.state.param.busy)
def indicator(busy):
    return "I'm busy" if busy else "I'm idle"
```

```python
def download_analysis_info_process(event):
    weekly_update_analysis_info_revenue()

button_weekly_update_analysis = pn.widgets.Button(name='Download Analysis Info')
button_weekly_update_analysis.on_click(download_analysis_info_process)

def download_balance_sheet_process(event):
    weekly_update_balance_sheet()
    
button_weekly_update_balance_sheet = pn.widgets.Button(name='Download Balance Sheet')
button_weekly_update_balance_sheet.on_click(download_balance_sheet_process)

def download_cash_flow_process(event):
    weekly_update_cash_flow()

button_weekly_update_cash_flow = pn.widgets.Button(name='Download Balance Sheet')
button_weekly_update_cash_flow.on_click(download_cash_flow_process)

def download_historical_price_process(event):
    weekly_update_historical_price()
    
button_weekly_update_historical_price = pn.widgets.Button(name='Download Historical Price')
button_weekly_update_historical_price.on_click(download_historical_price_process)

def download_income_statement_process(event):
    weekly_update_income_statement()

button_weekly_update_income_statement = pn.widgets.Button(name='Download Income Statement')
button_weekly_update_income_statement.on_click(download_income_statement_process)   
```

```python
def valuation_sp500_processing(event):
    input_config_dir = INPUT_DIR
    input_config_file = 'input.ymal'
    valuation_multiple_stock(input_config_dir, input_config_file) 
    
button_valuation_sp500 = pn.widgets.Button(name='Valuation SP500')
button_valuation_sp500.on_click(valuation_sp500_processing)

```

```python
tabs = pn.Tabs()
# Add a tab
tabs.append(('Download Data', pn.Column(pn.Row(button_weekly_update_analysis, indicator),
                                        pn.Row(button_weekly_update_balance_sheet, indicator),
                                        pn.Row(button_weekly_update_cash_flow, indicator),
                                        pn.Row(button_weekly_update_historical_price, indicator),
                                        pn.Row(button_weekly_update_income_statement, indicator)
                                       )))

tabs.extend([
    ('Valuation', pn.Row(button_valuation_sp500, indicator)),
])

tabs
```
